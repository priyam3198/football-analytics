Let's start with very basic graphs related to Liverpool's matches in 2015/16 season

# **WIN AND LOSS PERCENTAGES**

```python
total_matches = len(lvr)
win_home = len(lvr_home[lvr_home['home_score'] > lvr_home['away_score']])
win_away = len(lvr_away[lvr_away['away_score'] > lvr_away['home_score']])
lose_home = len(lvr_home[lvr_home['away_score'] > lvr_home['home_score']])
lose_away = len(lvr_away[lvr_away['home_score'] > lvr_away['away_score']])
draws = total_matches - (win_home + win_away + lose_home + lose_away)
```

```python
percentage = [win_home / total_matches * 100, win_away / total_matches * 100, draws/total_matches * 100, lose_home / total_matches * 100, lose_away / total_matches * 100,]
```

```python
labels = ['Home Win', 'Away Win', 'Draws', 'Lose Home', 'Lose Away']
colors = ['#99b3ff', '#66b3ff', '#CCCCCC', '#ff6666', '#ff9999']
explode = (0, 0,0, 0, 0)
```

```python
plt.figure(figsize=(8, 8))
plt.pie(percentage, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, explode=explode)
plt.title("Match Distribution")
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x=labels, y=percentage, palette=colors)
plt.xlabel("Outcome")
plt.ylabel("Percentage")
plt.title("Match Distribution")
```

# **GOAL SCORING ANALYSIS**

```python
total_home_goals = lvr_home['home_score'].sum()
total_home_conceded = lvr_home['away_score'].sum()

total_away_goals = lvr_away['away_score'].sum()
total_away_conceded = lvr_away['home_score'].sum()
```

```python
cumulative_sum_home_scored =lvr_home['home_score'].cumsum()
cumulative_sum_home_conceded = lvr_home['away_score'].cumsum()

cumulative_sum_away_scored = lvr_away['away_score'].cumsum()
cumulative_sum_away_conceded = lvr_away['home_score'].cumsum()

```

```python
x_values = range(1, len(cumulative_sum_home_scored) + 1)
fig, ax = plt.subplots(1,2,figsize=(12, 6))

sns.lineplot(x=x_values, y=cumulative_sum_home_scored, label='Home Goals Scored', ax=ax[0])
sns.lineplot(x=x_values, y=cumulative_sum_home_conceded, label='Home Goals Conceded', ax=ax[0])

x_ticks = range(min(x_values), max(x_values) + 1, 2)
y_ticks = range(min(cumulative_sum_home_scored), max(cumulative_sum_home_scored) + 1, 4)
ax[0].set_xticks(x_ticks)
ax[0].set_yticks(y_ticks)

plt.xlabel('Matches')
plt.ylabel('Goals')
plt.title('Home Goals Scored & Conceded')
plt.legend()

sns.lineplot(x=x_values, y=cumulative_sum_away_scored, label='Away Goals Scored', ax=ax[1])
sns.lineplot(x=x_values, y=cumulative_sum_away_conceded, label='Away Goals Conceded', ax=ax[1])

x_ticks_away = range(min(x_values), max(x_values) + 1, 2)
y_tick_away = range(min(cumulative_sum_away_scored), max(cumulative_sum_away_scored) + 1, 4)
ax[1].set_xticks(x_ticks_away)
ax[1].set_yticks(y_tick_away)

plt.xlabel('Matches')
plt.ylabel('Goals')
plt.title('Away Goals Scored & Conceded')
plt.legend()

plt.show()
```
